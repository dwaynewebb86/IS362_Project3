Import the necessary modules

```python
import sqlite3
import pandas as pd
```

Calling the database

```python
conn = sqlite3.connect("Chinook_Sqlite.sqlite")
```

Now that the database I will query it to provide the Customers’ Last
Name and First Name, and each customer’s purchased Track names and
Album Titles in a pandas DataFrame sorted by Customer LastName then Customer FirstName

```python
query = """
SELECT 
    c.LastName,
    c.FirstName,
    t.Name    AS Name,
    a.Title   AS Title
FROM Customer c
JOIN Invoice i      ON c.CustomerId = i.CustomerId
JOIN InvoiceLine il ON i.InvoiceId  = il.InvoiceId
JOIN Track t        ON il.TrackId   = t.TrackId
JOIN Album a        ON t.AlbumId    = a.AlbumId
ORDER BY 
    c.LastName,
    c.FirstName
"""
df = pd.read_sql(query, conn)
```
